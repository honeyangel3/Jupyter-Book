```python
# Perform Standard Imports

import torch
import numpy as np
import sys
```

```python
# Create Function

def set_seed(seed: int) -> None:
    torch.manual_seed(seed)
    np.random.seed(seed)
```

```python
# Create a Numpy Array

set_seed(42)
arr = np.random.randint(low=0, high=5, size=6)
```

```python
# Create tensor

x = torch.from_numpy(arr)
print(x)
```

```python
# Change dtype


dt = x.type(torch.int64)
print(dt)
print(dt.dtype)
```

```python
# Reshape 

x = torch.arange(6).reshape(3,2)
print(x)
```

```python
# Return right hand column

x[:,1]
```

```python
# Return a tensor of square values

x_squared = x.pow(2)
print(x_squared)
```

```python
# Create a tensor y

set_seed(42)
y = torch.randint(low=0, high=5, size=(2,3), dtype= torch.int64)
y
```

```python
# Find the Matrix Product of x and y
print(torch.mm(x,y))
```
